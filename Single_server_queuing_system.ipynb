```python
# Name: Rutuja Kolhe
# MIS: 111807018
# Batch: B
# Subject: Simulation Tools Lab

# Lab 1
#Single Server Queuing System Simulation


#Importing the necessary libraries
import pandas as pd
import numpy as np
import random

avg_wait_time=0
prob_cust_waits=0
idle_server_time=0
avg_service_time=0
avg_time_betn_intervals=0
avg_wait_time=0
avg_waittime_system=0

for m in range(10):

    print("Simulation results for iteration number ",m+1)
    
    #Number of customers
    n=6


    #Defining variables
    cust=np.zeros((n,),dtype=int)           # Customers
    inter_arrival=np.zeros((n,),dtype=int)  # Inter-arrival time
    arrival_t=np.zeros((n,),dtype=int)      # Arrival time
    service_t=np.zeros((n,),dtype=int)      # Service time

    t_serv_begin=np.zeros((n,),dtype=int)   # Time service begins
    t_cust_waits=np.zeros((n,),dtype=int)   # Time customer waits in queue
    t_serv_ends=np.zeros((n,),dtype=int)    # Time service ends
    t_cust_spends=np.zeros((n,),dtype=int)  # Time customer spends in system 
    idle_t_server=np.zeros((n,),dtype=int)  # Idle time of the server


    #Inputting the distribution for random number generation
    servicetime_list = [1,2,3,4,5,6]
    distribution_s = [0.1,0.15,0.25,0.25,0.1,0.15]

    interarrivaltime_list = [1,2,3,4,5,6,7,8]
    distribution_i = [0.25,0.15,0.125,0.125,0.1,0.1,0.05,0.1]


    # Assigning Customer number and generating random number
    for i in range(n):
        cust[i]=i+1
        rand_num1=random.choices(interarrivaltime_list, distribution_i)
        rand_num2=random.choices(servicetime_list, distribution_s)
        inter_arrival[i]=rand_num1[0]
        service_t[i]=rand_num2[0]

    inter_arrival[0]=0
    arrival_t[0]=0

    for i in range(n):
        arrival_t[i]=arrival_t[i-1]+inter_arrival[i]


    # Calculating Time Service Begins and Time Customer Waits
    t_serv_begin[0]=0
    t_cust_waits[0]=0
    idle_t_server[0]=0

    for i in range(1,n):
        t_serv_begin[i]=max(arrival_t[i],t_serv_begin[i-1]+service_t[i-1])
        t_cust_waits[i]=t_serv_begin[i]-arrival_t[i]


    # Calculating Service ends time and Customer spends time
    for i in range(n):
        t_serv_ends[i]=t_serv_begin[i]+service_t[i]
        t_cust_spends[i]=service_t[i]+t_cust_waits[i]


    # Calculating Idle time of server
    for i in range(1,n):
        idle_t_server[i]=max(0, arrival_t[i]-t_serv_ends[i-1])


    # Average waiting time for a customer
    total_waiting_t_cust=sum(t_cust_waits)
    avg_waiting_t_cust=total_waiting_t_cust/len(cust)


    # Probability that a customer has to wait
    num_cust_wait=np.count_nonzero(t_cust_waits)
    prob_cust_wait=num_cust_wait/len(cust)


    # The fraction of idle time of Server
    total_idle_serv_t=sum(idle_t_server)
    frac_idle_t_serv=total_idle_serv_t/t_serv_ends[n-1]


    # The average service time
    total_service_t=sum(service_t)
    avg_service_t=total_service_t/len(cust)


    # The average time between arrivals
    total_int_arrival=sum(inter_arrival)
    avg_int_arrival=total_int_arrival/(len(arrival_t)-1)


    # The average waiting time for those who are waiting
    avg_wait_for_those_cust=total_waiting_t_cust/num_cust_wait


    # The average waiting time a customer spends
    total_t_cust_spends=sum(t_cust_spends)
    avg_t_cust_spends=total_t_cust_spends/len(cust)


    from prettytable import PrettyTable
    x = PrettyTable()
    column_names = ['Customer','IAT','AT','ST','TSB','TCWQ','TSE','TCSS','System Ideal']
    data = [cust,inter_arrival,arrival_t,service_t,t_serv_begin,t_cust_waits,t_serv_ends,t_cust_spends,idle_t_server]
    length = len(column_names)
    for i in range(length):
      x.add_column(column_names[i],data[i])
    print(x)
    
    avg_wait_time+=avg_waiting_t_cust
    prob_cust_waits+=prob_cust_wait
    idle_server_time+=frac_idle_t_serv
    avg_service_time+=avg_service_t
    avg_time_betn_intervals+=avg_int_arrival
    avg_wait_time+=avg_wait_for_those_cust
    avg_waittime_system+=avg_t_cust_spends
    
    
    # Printing asked question values
    print("\nAverage waiting time for customer: ","{:.3f}".format(avg_waiting_t_cust))
    print("Probability that a customer has to wait is: ","{:.3f}".format(prob_cust_wait))
    print("Fraction of the idle time of a server: ","{:.3f}".format(frac_idle_t_serv))
    print("Average Service Time: ","{:.3f}".format(avg_service_t))
    print("Average time between arrivals: ","{:.3f}".format(avg_int_arrival))
    print("Average waiting time for those who wait: ","{:.3f}".format(avg_wait_for_those_cust))
    print("Average waiting time a customer spends: ","{:.3f}".format(avg_t_cust_spends))
    print("\n")

    df=pd.DataFrame({"Customer":cust,"Inter-Arrival Time":inter_arrival, "Arrival Time":arrival_t, "Service Time":service_t,"Time Service begins":t_serv_begin,"Time Customer Waits":t_cust_waits,"Time Service Ends":t_serv_ends,"Time Customer Spends":t_cust_spends,"Idle Time of Server":idle_t_server})
    if m==0:filename="iteration1.csv"
    elif m==1:filename="iteration2.csv"
    elif m==2:filename="iteration3.csv"
    elif m==3:filename="iteration4.csv"
    elif m==4:filename="iteration5.csv"
    elif m==5:filename="iteration6.csv"
    elif m==6:filename="iteration7.csv"
    elif m==7:filename="iteration8.csv"
    elif m==8:filename="iteration9.csv"
    elif m==9:filename="iteration10.csv"
    df.to_csv(filename,index=False)

print("\nThe average answers of all the iterations are:")
print("\n1. The average waiting time for a customer:",avg_wait_time/10)
print("\n2. The probability that a customer has to wait in a queue:",prob_cust_waits/10)
print("\n3. The fraction of the idle time of a server:",idle_server_time/10)
print("\n4. The average service time:",avg_service_time/10)
print("\n5. The average time between arrivals:",avg_time_betn_intervals/10)
print("\n6. The average waiting time for those who are waiting:",avg_wait_time/10)
print("\n7. The average waiting time a customer spends in the system:",avg_waittime_system/10)

```
